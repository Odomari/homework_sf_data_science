```python
import pandas as pd
import seaborn as sns

import sys
sys.path.append('/Users/familyafanasyevy/Documents/Денис/Skillfactory/vulnerable_populations/additional_functions')
from functions import to_modify_dataframe, transform_to_changes
```

##### Прочитаем и преобразуем данные об обороте розничной торговле на душу населения.

```python
retail_turnover_per_capita = pd.read_excel(
    io='../social_russia_data/retail_turnover_per_capita_2000_2021.xls', sheet_name=0, header=2, index_col=0, usecols='A, C:X', skiprows=[3]
)
retail_turnover_per_capita = to_modify_dataframe(retail_turnover_per_capita)
retail_turnover_per_capita = retail_turnover_per_capita.rename(lambda arg: f'Оборот розничной торговли на душу населения в {arg} году, руб.', axis=1)
retail_turnover_per_capita.head()
```

##### Посмотрим корреляцию признаков.

```python
sns.heatmap(retail_turnover_per_capita.corr(), annot=True);
```

##### Данные очень сильно коррелированы. Поэтому рассмотрим динамику изменения оборота розничной торговли.

```python
retail_turnover_per_capita_changes = transform_to_changes(retail_turnover_per_capita)
retail_turnover_per_capita_changes = retail_turnover_per_capita_changes.rename(
    lambda arg: f"Динамика оборота розничной торговли на душу населения на {arg.split(' ')[7]} г., в %", axis=1
)
retail_turnover_per_capita_changes.head()
```
