##### Теперь, когда мы прочитали и предобработали данные, объединим все таблицы в единый датафрейм.

```python
merged_df = population_changes
for dataframe in dataframes_for_merge:
    merged_df = merged_df.merge(
        right=dataframe,
        how='inner',
        left_index=True,
        right_index=True
    )
merged_df.head()
```

##### Рассчитаем корреляцию всех признаков между собой. Если найдётся какая-либо пара признаков, коэффициент корреляции между которыми больше 0.9 по абсолютной величине, то удалим одного представителя этой пары из получившегося датафрейма. Удалять будем тот признак из пары, сумма коэффициентов корреляции которого с остальными признаками, наименьшая.

```python
corr_pairs, corr_pair_min = to_evaluate_corr_pairs(merged_df, 0.9)

matrix_corr = merged_df.corr()
features_for_drop = set()
for pair in corr_pairs:
    if matrix_corr[pair[0]].sum() < matrix_corr[pair[1]].sum():
        features_for_drop.add(pair[0])
    else:
        features_for_drop.add(pair[1])

print(f'Число признаков, которые будут удалены - {len(features_for_drop)}.')
merged_df = merged_df.drop(columns=features_for_drop)
print(f'Число признаков, оставшихся после удаления - {merged_df.shape[1]}.')
```
